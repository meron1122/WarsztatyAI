## Neuron McCullocha-Pittsa

Neuron McCullocha-Pittsa jest prb implementacji matematycznej zachowania neuronu w mzgu.

Jego schemat wyglda nastpujco:
obraz.png

Neuron ten przyjmuje na wejcie okrelon ilo cech, mnoy kad z cech przez okrelon wag, oraz przetwarza funkcj aktywacji, za ktr neutron przewodzi(wystawia 1), lub nie aktywuje si(wystawia 0). Cakiem podobnie jak ludzkie neurony. Dodatkowo neutron posiada wejcie obciajce tzw. Bias, ktry pozwala przesun zakres dziaania neuronu.

Reasumujc, matematycznie nasz neutron wyglda tak:

$y=F(\sum x_{n}*w_{n})$

Gdzie:
*   $y$ to wyjcie neutronu
*   $F$ to funkcja aktywacji
*   $x_{n}$ to poszczeglne wejcie
*   $w_{n}$ to poszczeglna waga


Teraz, jak ju mamy wzr, atwo bdzie nam stworzy neutron w kodzie

# Proces uczenia
Uczenie jest procesem, w ktrym dobieramy wartoci poszczeglnych wag, z gry okrelonym algorytmem. Uczenie jest najwaniejszym procesem - to dziki niemu nasza sie zaczyna zachowywa si *inteligentnie*.

Najprostszym algorytmem jest tzw. algorytm uczenia perceptronu(Simple perceptron learning algorithm, **SPLA**), ktry wyglda nastpujco:



1. Losujemy wagi z przedziau od (0,1)
2. Ze zbioru danych uczcych wybieramy losowe dane wejciowe $P$ oraz oczekiwan odpowied $T$
3. Obliczamy wynik dziaania sieci $O$
4. Obliczamy bd $Err=O-T$
5. Jeeli $Err = 0$ czyli przypadek poprawnie sklasyfikowany -> skocz do punktu 2
6. Jeeli nie, aktualizujemy kad z wag wg. wzoru $w_i = w_i + lr * Err * P_i$.
Aktualizujemy rwnie bias wg. wzoru $bias = bias + lr * Err$
7. Algorytm zaptlamy tak dugo, jak nie osigniemy satysfakcjonojcego wyniku

$lr$ ma sta uczenia, ktr przyjmujemy jako dane wejciowe do algorytmu

```python
#@title Rozwizanie

class Neuron:
  def __init__(self, input_size):
    self.weights = np.random.random(input_size)
    #bias
    self.bias = 1

  # funkcja aktywacji tzw. funkcja progowa
  def __activate(self, value):
    if value >= 0:
      return 1
    else:
      return 0

  def predict(self,inputs):
    sum = np.sum(self.weights * inputs)
    return self.__activate(sum + self.bias)

  def train(self, epochs, inputs,outputs, lerning_rate):
    #Zaptlenie algorytmu tak dugo jak uytkownik zdefiniuje
    for _ in range(epochs):
      #Kada epoka skada si z  tylu iteracji ile przypadkow
      for _ in range(inputs.shape[0]):
          #wylosuj index przypadku
          ri = np.random.randint(0,inputs.shape[0])
          #Wybierzy przypadek wg. losowego indexu, oraz jego wyjcie
          random_input = inputs[ri]
          random_output = outputs[ri]

          #oblicz bd
          error =  random_output - self.predict(random_input)

          if error != 0:
            #aktualizacja wag
            for j in range(len(self.weights)):
              self.weights[j] += error * random_input[j] * learning_rate

            self.bias += learning_rate * error
```

```python
##########
#### Wywoanie perceptronu
##########
#Czy ta tabelka prawdy nie wyglda znajomo? :)
inputs = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
outputs = np.array([
    1,
    1,
    1,
    0
])
epochs = 1000
learning_rate = 0.001

neuron = Neuron(2)
neuron.train(epochs,inputs,outputs,learning_rate)

for input in inputs:
  print(f"Dla wejcia:{input}, Neuron przewidzia warto: {neuron.predict(input)}")
```
