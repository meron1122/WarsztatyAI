# Perceptron

W 1957 Frank Rosenblatt prezentuje model najprostszej sieci neuronowej - peceptronu. Sieci, która zbudowana jest z jednego lub wielu neuronów McCullocha-Pittsa.  Zrewolucjonizowało to świat uczenia maszynowego, i dało początek rozwiązywaniu problemów klasyfikacji binarnej.

## Neuron McCullocha-Pittsa

Neuron McCullocha-Pittsa jest próbą implementacji matematycznej zachowania neuronu w mózgu.

Jego schemat wygląda następująco:
obraz.png

Neuron ten przyjmuje na wejście określoną ilość cech, mnoży każdą z cech przez określoną wagę, oraz przetwarza funkcją aktywacji, za którą neutron przewodzi(wystawia 1), lub nie aktywuje się(wystawia 0). Całkiem podobnie jak ludzkie neurony. Dodatkowo neutron posiada wejście obciążające tzw. Bias, który pozwala przesunąć zakres działania neuronu.

Reasumując, matematycznie nasz neutron wygląda tak:

$y=F(\sum x_{n}*w_{n})$

Gdzie:
*   $y$ to wyjście neutronu
*   $F$ to funkcja aktywacji
*   $x_{n}$ to poszczególne wejście
*   $w_{n}$ to poszczególna waga


Teraz, jak już mamy wzór, łatwo będzie nam stworzyć neutron w kodzie 🙂

# Proces uczenia
Uczenie jest procesem, w którym dobieramy wartości poszczególnych wag, z góry określonym algorytmem. Uczenie jest najważniejszym procesem - to dzięki niemu nasza sieć zaczyna zachowywać się *inteligentnie*.

Najprostszym algorytmem jest tzw. algorytm uczenia perceptronu(Simple perceptron learning algorithm, **SPLA**), który wygląda następująco:



1. Losujemy wagi z przedziału od (0,1)
2. Ze zbioru danych uczących wybieramy losowe dane wejściowe $P$ oraz oczekiwaną odpowiedź $T$
3. Obliczamy wynik działania sieci $O$
4. Obliczamy błąd $Err=O-T$
5. Jeżeli $Err = 0$ czyli przypadek poprawnie sklasyfikowany -> skocz do punktu 2
6. Jeżeli nie, aktualizujemy każdą z wag wg. wzoru $w_i = w_i + lr * Err * P_i$.
Aktualizujemy również bias wg. wzoru $bias = bias + lr * Err$
7. Algorytm zapętlamy tak długo, jak nie osiągniemy satysfakcjonojącego wyniku

$lr$ małą stałą uczenia, którą przyjmujemy jako dane wejściowe do algorytmu

```python
#@title Rozwiązanie

class Neuron:
  def __init__(self, input_size):
    self.weights = np.random.random(input_size)
    #bias
    self.bias = 1

  # funkcja aktywacji tzw. funkcja progowa
  def __activate(self, value):
    if value >= 0:
      return 1
    else:
      return 0

  def predict(self,inputs):
    sum = np.sum(self.weights * inputs)
    return self.__activate(sum + self.bias)

  def train(self, epochs, inputs,outputs, lerning_rate):
    #Zapętlenie algorytmu tak długo jak użytkownik zdefiniuje
    for _ in range(epochs):
      #Każda epoka składa się z  tylu iteracji ile przypadkow
      for _ in range(inputs.shape[0]):
          #wylosuj index przypadku
          ri = np.random.randint(0,inputs.shape[0])
          #Wybierzy przypadek wg. losowego indexu, oraz jego wyjście
          random_input = inputs[ri]
          random_output = outputs[ri]

          #oblicz błąd
          error =  random_output - self.predict(random_input)

          if error != 0:
            #aktualizacja wag
            for j in range(len(self.weights)):
              self.weights[j] += error * random_input[j] * learning_rate

            self.bias += learning_rate * error
```

```python
##########
#### Wywołanie perceptronu
##########
#Czy ta tabelka prawdy nie wygląda znajomo? :)
inputs = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
outputs = np.array([
    1,
    1,
    1,
    0
])
epochs = 1000
learning_rate = 0.001

neuron = Neuron(2)
neuron.train(epochs,inputs,outputs,learning_rate)

for input in inputs:
  print(f"Dla wejścia:{input}, Neuron przewidział wartość: {neuron.predict(input)}")
```
